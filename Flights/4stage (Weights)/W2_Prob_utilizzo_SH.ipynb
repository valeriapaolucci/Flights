```python
clean = merged_2[(merged_2['minute_start']<merged_2['entry']+30) & (merged_2['entry']<=merged_2['minute_end']+30)]
#clean_prova = merged_2[(merged_2['minute_start']<=merged_2['entry']+29) & (merged_2['entry']<=merged_2['minute_end']+30)]
#print(len(clean)==len(clean_prova))
clean
```

```python
def f_list(row):
    return len_dict[row['O_D_A']]
```

```python
clean['number'] = clean.apply(f_list, axis=1)
```

```python
clean
```

```python
#max(clean['number'])
#clean[clean['number']>60]
#len_dict['RJTT_EDDF_B744']
```

```python
clean['bool'] = clean['entry'] + 30 >= clean['minute_end']
```

```python
def f_compute(row):
    if row['bool'] == True:
        value = row['minute_end'] - (row['entry'] - 30) + 1
    else:
        value = row['entry'] + 30 - row['minute_start']
    return max(0,value)
```

```python
clean['overlap'] = clean.apply(f_compute, axis=1)
```

```python
#max(clean['overlap'])
# clean[clean['overlap']>60]
```

```python
clean['prob'] = clean['overlap']/60
```

```python
clean['weighted'] = clean['prob']/clean['number']
```

```python
clean
```

```python
clean2 = clean.groupby('sector_hours')['weighted'].sum()

clean2 = pd.DataFrame({'sector_hours':clean2.index, 'Exp_freq':clean2.values})

clean2
```

```python
sectorhours2 = sectorhours.merge(clean2, how='left', on='sector_hours')
sectorhours2
```

```python
sectorhours2['load_factor'] = sectorhours2['Exp_freq']/sectorhours2['capacity']
sectorhours2 = sectorhours2[['sector_hours', 'icao', 'minute_start', 'minute_end', 'Exp_freq', 'capacity', 'load_factor']]
sectorhours2
```

# 18    
    
    Con l'istante di partenza fissato, era necessario che:
    1) entry - start >= 0   ovvero:   start - entry <= 0     ovvero: start <= entry
    2) end - entry >= 0     ovvero:   end >= entry           ovvero: entry <= end
    
    Con la flessibilitÃ  di piÃ¹/meno 30 minuti, basta che:
    1) (entry + 30) - start >= 0     ovvero: start - 30 <= entry     ovvero: start <= entry + 30
    2) end - (entry - 30) >= 0       ovvero: end + 30 > entry        ovvero: entry <= end + 30

    B) settore-ora attivo dal min 0 al min 60
