```python
sectorhours2 = sectorhours1[sectorhours1['sector_hours'].str.endswith('E')]
sectorhours2.reset_index(drop=True, inplace=True)
sectorhours2
```

```python
sectorhours2 = sectorhours2.merge(entry, how='left', left_on='sector_hours', right_on='Var1')
sectorhours2
```

```python
sectorhours2 = sectorhours2.drop(columns=['Var1'])
sectorhours2
```

```python
sectorhours2['load_factor'] = sectorhours2['Freq']/sectorhours2['capacity']
sectorhours2 = sectorhours2[['sector_hours', 'icao', 'minute_start', 'minute_end', 'Freq', 'capacity', 'load_factor']]
sectorhours2
```
