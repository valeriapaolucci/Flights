```python
def common_element(list1, list2): 
    for x in list1:
        for y in list2:
            if x == y:
                return x  
    return False
```

```python
l1 = ['ciao', 'sole', 'luna']
l2 = ['letto', 'cibo', 'casa']
l3 = []
common_element(l1,l3)
```

```python
for i in range(0,len(mydata)):
    # for each row, consider only the entries whose column index is smaller than i (the current row index)
    print('running', i)
    for ii in range(i):
        tmp = (common_element(mydata['sector_hours'].loc[i], mydata['sector_hours'].loc[ii]))
        if tmp != False:
            myloc = lf[lf['sector_hours'] == str(tmp)].index[0]
            if (lf['load_factor'].loc[myloc]>=0.8):
                # undirected graph
                AM3[i,ii] = AM3[ii,i] = 1
            
```
