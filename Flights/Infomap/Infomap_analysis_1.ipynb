```python
import pandas as pd
import numpy as np
np.set_printoptions(threshold=100000)
```

```python
flights = pd.read_pickle('C:\\Users\\utente\\Documents\\DSSC_IIyear\\Voli\\Dataframe\\S1_flights_sectorhours.pkl')
flights
```

```python
dictFlights = {i:flights['flight'].loc[i-1] for i in range(1,len(flights)+1)}
dictFlights
```

```python
data = pd.read_csv('infomap_output_W.txt', sep=" ", header=None)
data.columns = ["Flight", "Cluster", "drop"]
data
```

```python
min(data['Flight'])
data
```

```python
for i in range(0, len(data)):
    data['Flight'].loc[i] = dictFlights[(data['Flight'].loc[i])]
```

```python
mydata2 = data.groupby(['Cluster'])
mydata2 = mydata2['Flight'].unique()
mydata2 = pd.DataFrame({'CLUSTER':mydata2.index, 'FLIGHTS':mydata2.values})
mydata2
```

```python
mydata2['SIZE']=0
for i in range(0,len(mydata2)):
    mydata2['SIZE'].loc[i] = len(mydata2['FLIGHTS'].loc[i])
mydata2
```

```python
mydata2[mydata2['SIZE']>1]
```

```python
singletons = mydata2[mydata2['SIZE']==1]
singletons
```

```python
singletons.reset_index(drop=True, inplace=True)
singletons['CLUSTER'] = 'None'
singletons
```

```python
for i in range(0,len(singletons)):
    singletons['FLIGHTS'].loc[i] = ((singletons['FLIGHTS'].loc[i]).tolist())[0]
singletons
```

```python
type(singletons['FLIGHTS'].loc[1])
```

```python
singletons = singletons.groupby(['CLUSTER'])
singletons = singletons['FLIGHTS'].unique()
singletons = pd.DataFrame({'CLUSTER':singletons.index, 'FLIGHTS':singletons.values})
singletons['SIZE'] = len(singletons['FLIGHTS'].loc[0])
singletons
```

```python
clusters = mydata2[mydata2['SIZE']!=1]
clusters
```

```python
clusters['CLUSTER'] = range(0,len(clusters))
clusters
```

```python
a = clusters['SIZE'].sum()
b = singletons['SIZE'].loc[0]
c = 29917
print(a, b, c, a+b==c)
```

```python
infomap_clusters_1 = pd.concat([clusters, singletons])
infomap_clusters_1
```

```python
infomap_clusters_1.reset_index(drop=True, inplace=True)
infomap_clusters_1
```

```python
infomap_clusters_1['SIZE'].sum()
```

```python
infomap_clusters_1.to_csv('C:\\Users\\utente\\Documents\\DSSC_IIyear\\Voli\\Infomap\\infomap_clusters_W.csv', header=True, index=False)
```
