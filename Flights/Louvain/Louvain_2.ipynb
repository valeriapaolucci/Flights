```python
s  = pd.Series(partition,index=partition.keys())
s
```

```python
max(s)
```

```python
mydata = pd.DataFrame(s)
mydata['idx'] = mydata.index
mydata = mydata[['idx', 0]]
mydata
```

```python
mydata2 = mydata.groupby([0])
mydata2 = mydata2['idx'].unique()
mydata2 = pd.DataFrame({'MOD_CLASS':mydata2.index, 'FLIGHTS':mydata2.values})
mydata2
```

```python
mydata2['SIZE']=0
for i in range(0,len(mydata2)):
    mydata2['SIZE'].loc[i] = len(mydata2['FLIGHTS'].loc[i])
mydata2
```

```python
singletons = mydata2[mydata2['SIZE']==1]
singletons
```

```python
singletons.reset_index(drop=True, inplace=True)
singletons['MOD_CLASS'] = 'None'
singletons
```

```python
for i in range(0,len(singletons)):
    singletons['FLIGHTS'].loc[i] = ((singletons['FLIGHTS'].loc[i]).tolist())[0]
singletons
```

```python
type(singletons['FLIGHTS'].loc[1])
```

```python
singletons = singletons.groupby(['MOD_CLASS'])
singletons = singletons['FLIGHTS'].unique()
singletons = pd.DataFrame({'MOD_CLASS':singletons.index, 'FLIGHTS':singletons.values})
singletons['SIZE'] = len(singletons['FLIGHTS'].loc[0])
singletons
```

```python
clusters = mydata2[mydata2['SIZE']!=1]
clusters
```

```python
clusters['MOD_CLASS'] = range(0,len(clusters))
clusters
```

```python
a = clusters['SIZE'].sum()
b = singletons['SIZE'].loc[0]
c = len(flights)
print(a, b, c, a+b==c)
```

```python
modularity_2 = pd.concat([clusters, singletons])
modularity_2
```

```python
modularity_2.reset_index(drop=True, inplace=True)
modularity_2
```
