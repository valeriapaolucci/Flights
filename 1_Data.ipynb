# 1
Import the tools.

```python
import pandas as pd
import numpy as np
import copy
```

# 2
Read the data.

```python
flights = pd.read_csv('flights.csv')
sectorhours = pd.read_csv('sectorhours.csv')
trajectories = pd.read_csv('trajectories.csv')
```

# 3
Print the data.

```python
print(flights)
print (sectorhours)
print (trajectories)
```

# 4
Drop unnecessary columns in "flights".

```python
flights = flights.drop(columns=['O_D_A', 'dep_time'])
flights
```

# 5
Consider only the sectorhours with role 'E'. <br>Reset the index.

```python
sectorhours = sectorhours[sectorhours['role']=='E']
sectorhours.reset_index(drop=True, inplace=True)
sectorhours
```

# 6
Merge "trajectories" with "flights".

E' indifferente applicare merge con modalità 'inner' o 'right': tutte le traiettorie in flights hanno un corrispondente in trajectories (il viceversa non è vero, come si potrebbe verificare applicando la modalità 'left': ci sono traiettorie in trajectories che non sono state volate il 1/9/2017).

```python
merged = trajectories.merge(flights, how='right', left_on='centroid', right_on='original_trajectory')
merged
```

# 7
Drop unnecessary columns.
<br>Reorder the columns.
<br>Rename the columns.

```python
merged = merged.drop(columns=['eobt', 'time_entry', 'original_trajectory'])
merged = merged[['uuid', 'centroid', 'icao', 'dep_time_minutes', 'time_to_sector_from_origin']]
merged.columns = ['flight', 'trajectory', 'icao', 'dep_time', 'time_to_sector']
merged
```

# 8
Create a new column 'entry', describing the time a flight enters a sector-hour.
<br>The modulo operation is needed since we might exceed midnight (60x24=1440).

```python
merged['entry'] = (merged['dep_time'] + merged['time_to_sector'])%1440
merged
```

# 9
Check that the maximum value of the column 'entry' doesn't exceed 1440.

```python
print ("Is the maximum value of the column 'entry' smaller than 1440?")
print (max(merged['entry'])<1440)
print("max(merged['entry']) =", max(merged['entry']))
```

# 10
Merge "merged" with "sectorhours", based on 'icao'.

```python
merged_2 = merged.merge(sectorhours, how='left', left_on='icao', right_on='icao')
merged_2
```

The modality 'left' keeps all the sectorhours present in "merged", even when they don't have any corrispondence in "sectorhours" (the relevant entries are set to 'Nan').
The modality 'inner', instead, would keep only the sectorhours in "merged" having an icao which is also present in "sectorhours", thus leaving out a number of flights.

# 11
Keep only the sectorhours having a 'minute_start' not greater than 'entry' and a 'minute_end' not smaller than 'entry'.
<br>Compare the number of flights included in "merged_2" and "clean2" respectively.

```python
clean = merged_2[merged_2['minute_start']<=merged_2['entry']]
clean2 = clean[clean['entry']<=clean['minute_end']]
```

```python
l1 = merged_2['flight'].unique()
l2 = clean2['flight'].unique()
print("len(l1): ", len(l1))
print("len(l2): ", len(l2))
diff = list(set(l1) - set(l2))
print("len(diff): ", len(diff))
```

The number of flights being equal to 29858, tells us that 59 flights are left out.
This is due to the fact that the merge modality 'left' keeps all the sectorhours present in "merged", even when they don't have any corrispondence in "sectorhours"; the missing entries are then set to 'NaN', but the inequalities of step 11 leave out the rows corresponding to such entries.
The 59 missing flights are included again at a later step (#14 and 15).

# 12
Drop unnecessary columns. <br>Reset the index.

```python
clean2 = clean2.drop(columns=['icao', 'time_to_sector', 'entry', 'role', 'minute_start', 'minute_end', 'capacity'])
clean2.reset_index(drop=True, inplace=True)
clean2
```

# 13
Collapse all the sectorhours associated to each flight into a list of sectorhours.
<br>A new dataframe is created, containing the two columns 'flight' and 'sector_hours'.

```python
clean3 = clean2.groupby('flight')
clean4 = clean3['sector_hours'].unique()

clean5 = pd.DataFrame({'flight':clean4.index, 'sector_hours':clean4.values})

clean5
```

# 14
Recover the 59 missing flights: create a new dataframe "flights2" selecting from "flights" only the rows having an 'uuid' that is present in the list "diff". <br>Drop the unnecessary columns. <br>Rename the column; add a new column named 'sector_hours', with empty lists as entries.
<br>Check that the lenght of this new dataframe is 59.

```python
flights2 = flights[flights['uuid'].isin(diff)]
flights2 = flights2.drop(columns=['original_trajectory', 'dep_time_minutes'])
flights2.columns = ['flight']
flights2['sector_hours'] = [[] for _ in range(len(flights2))]
    
print(len(flights2))    
flights2
```

# 15
Append "flights2" to "clean".
<br>Sort the complete dataframe accoridng to the column 'flight'.
<br>Reset the index.

```python
clean5 = clean5.append(flights2)
clean5.sort_values(by=['flight'], inplace=True)
clean5.reset_index(drop=True, inplace=True)
clean5
```

# 16
Export the dataframe as pickle (in order to preserve the type of the objects).

```python
clean5.to_pickle('flights_sectorhours.pkl')
```
