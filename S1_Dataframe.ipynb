# 5
Consider only the sectorhours with role 'E'. <br>Reset the index.

```python
sectorhours = sectorhours[sectorhours['role']=='E']
sectorhours.reset_index(drop=True, inplace=True)
sectorhours
```

# 6
Create a new column 'O_D_A' in trajectories, obtained from the first 14 characters of 'centroid'.
<br> Drop the unnecessary columns.
<br> Rename the columnns.
<br> Check that the entries of 'O_D_A' are all of lenght 14. 

```python
trajectories['O_D_A'] = trajectories['centroid'].str[0:14]
trajectories = trajectories.drop(columns=['eobt', 'time_entry'])
trajectories = trajectories[['O_D_A', 'centroid', 'icao', 'time_to_sector_from_origin']]
```

# 7
Merge "trajectories" with "flights".

E' indifferente applicare merge con modalità 'inner' o 'right': tutte le traiettorie in flights hanno un corrispondente in trajectories (il viceversa non è vero, come si potrebbe verificare applicando la modalità 'left': ci sono traiettorie in trajectories che non sono state volate il 1/9/2017).

```python
merged = trajectories.merge(flights, how='right', on = 'O_D_A')
merged
```

# 8
Drop unnecessary columns.
<br>Reorder the columns.
<br>Rename the columns.

```python
merged = merged[['uuid', 'O_D_A', 'centroid', 'icao', 'dep_time_minutes', 'time_to_sector_from_origin']]
merged.columns = ['flight', 'O_D_A', 'trajectory', 'icao', 'dep_time', 'time_to_sector']
merged
```

# 9
Create a new column 'entry', describing the time a flight enters a sector-hour.
<br>The modulo operation is needed since we might exceed midnight (60x24=1440).

```python
merged['entry'] = (merged['dep_time'] + merged['time_to_sector'])%1440
merged
```

# 10
Check that the maximum value of the column 'entry' doesn't exceed 1440.

```python
print ("Is the maximum value of the column 'entry' smaller than 1440?")
print (max(merged['entry'])<1440)
print("max(merged['entry']) =", max(merged['entry']))
```

# 11
Merge "merged" with "sectorhours", based on 'icao'.

```python
merged_2 = merged.merge(sectorhours, how='left', left_on='icao', right_on='icao')
merged_2
```

The modality 'left' keeps all the sectorhours present in "merged", even when they don't have any corrispondence in "sectorhours" (the relevant entries are set to 'NaN').
The modality 'inner', instead, would keep only the sectorhours in "merged" having an icao which is also present in "sectorhours", thus leaving out a number of flights.

# 12
Keep only the sectorhours having a 'minute_start' not greater than 'entry' +30 and an 'entry' not smaller than 'minute_end' +30.
<br>Compare the number of flights included in "merged_2" and "clean" respectively.
See #18 for the reasoning.

```python
clean = merged_2[(merged_2['minute_start']<=merged_2['entry']+30) & (merged_2['entry']<=merged_2['minute_end']+30)]
clean
```

```python
l1 = merged_2['flight'].unique()
l2 = clean['flight'].unique()
print("len(l1): ", len(l1))
print("len(l2): ", len(l2))
diff = list(set(l1) - set(l2))
print("len(diff): ", len(diff))
```

The number of flights being equal to 29872, tells us that 45 flights are left out.
This is due to the fact that the merge modality 'left' keeps all the sectorhours present in "merged", even when they don't have any corrispondence in "sectorhours"; the missing entries are then set to 'NaN', but the inequalities of step 12 leave out the rows corresponding to such entries.
The 45 missing flights are included again at a later step (#15 and 16).

# 13
Drop unnecessary columns. <br>Reset the index.

```python
clean2 = clean.drop(columns=['icao', 'time_to_sector', 'entry', 'role', 'minute_start', 'minute_end', 'capacity'])
clean2.reset_index(drop=True, inplace=True)
clean2
```

# 14
Collapse all the sectorhours associated to each flight into a list of sectorhours.
<br>A new dataframe is created, containing the two columns 'flight' and 'sector_hours'.

```python
clean3 = clean2.groupby('flight')
clean4 = clean3['sector_hours'].unique()

clean5 = pd.DataFrame({'flight':clean4.index, 'sector_hours':clean4.values})

clean5
```

# 15
Recover the 45 missing flights: create a new dataframe "flights2" selecting from "flights" only the rows having an 'uuid' that is present in the list "diff". <br>Drop the unnecessary columns. <br>Rename the column; add a new column named 'sector_hours', with empty lists as entries.
<br>Check that the lenght of this new dataframe is 45.

```python
flights2 = flights[flights['uuid'].isin(diff)]
flights2 = flights2.drop(columns=['O_D_A', 'dep_time_minutes'])
flights2.columns = ['flight']
flights2['sector_hours'] = [[] for _ in range(len(flights2))]
    
print(len(flights2))    
flights2
```

# 16
Append "flights2" to "clean".
<br>Sort the complete dataframe accoridng to the column 'flight'.
<br>Reset the index.

```python
clean5 = clean5.append(flights2)
clean5.sort_values(by=['flight'], inplace=True)
clean5.reset_index(drop=True, inplace=True)
clean5
```
